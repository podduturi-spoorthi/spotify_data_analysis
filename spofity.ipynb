```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
track = pd.read_csv('C:/Users/dell/Downloads/archive/tracks.csv')
track.head()
```

```python
pd.isnull(track).sum()
```

```python
track.info()
```

```python
#top 10 least popular songs

min_popularity = track.sort_values('popularity', ascending = True).head(10)
min_popularity
```

```python
#top 10 popularity songs
max_popularity = track.sort_values('popularity',ascending = False).head(10)
max_popularity
```

```python
track.describe()
```

```python
#top 10 songs with popularity >90

most_popular = track.query('popularity>90',inplace = False).sort_values('popularity',ascending = False).head(10)
most_popular
```

```python
track.set_index("release_date", inplace=True)
track.index = pd.to_datetime(track.index)
track.head()
```

```python
track[["artists"]].iloc[18]
```

```python
track[["name","popularity"]].iloc[2]
```

```python
#converting duration from millisec to sec

track['duration'] = track['duration_ms'].apply(lambda x: round(x/1000))
track.head(5)
```

```python
track.drop("duration_ms",inplace=True,axis=1)
```

```python
track.head(2)
```

```python
#creating vizz(correlation map)
corr_track =track.drop(['key','mode','explicit'],axis=1).corr(method='pearson')
plt.figure(figsize=(14,6))
heatmap = sns.heatmap(corr_track, annot=True, fmt='.1g', vmin=-1, vmax=1, center=0, cmap='inferno', linewidth=1, linecolor='Black')
heatmap.set_title("Correlation HeatMap Between Variable")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90)
```

```python
sample_track = track.sample(int(0.004*len(track)))
print(len(sample_track))
```

```python
#regression plot between loudness and sound

plt.figure(figsize = (10,6))
sns.regplot(data = sample_track,y='loudness',x='energy',color='c').set(title='loudness vs energy correlation')
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_track,y='popularity',x='acousticness',color='b').set(title='popularity vs acousticness correlation')
```

```python
track['dates'] = track.index.get_level_values('release_date')
track.dates = pd.to_datetime(track.dates)
years = track.dates.dt.year
```

```python
sns.displot(years,discrete=True,aspect=2,height=5,kind='hist').set(title ='no.of songs')
```

```python
total_dr = track.duration
fig_dims = (18,7)
fig, ax = plt.subplots(figsize= fig_dims)
fig = sns.barplot(x= years, y=total_dr,ax=ax,errwidth = False).set(title='year vs duration')
plt.xticks(rotation=90)
```

```python
total_dr = track.duration
sns.set_style(style='whitegrid')
fig, ax = plt.subplots(figsize = (10,5))
fig = sns.lineplot(x=years, y=total_dr, ax=ax).set(title="year vs duration")
plt.xticks(rotation=60)
```
